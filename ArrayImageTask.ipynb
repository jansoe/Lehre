```python
import sklearn.datasets
import matplotlib.pyplot as plt 
import numpy as np
```

## Intro

First we load a test image

```python
image = sklearn.datasets.load_sample_image('flower.jpg')
_ = plt.imshow(image)
```

The image is a numpy array

```python
type(image)
```

with 3 dimension

```python
image.shape
```

The first dimension is the height 427, the second the width 640 and the third dimension contains the 3 color channels.
The different channel represents the amount of the different colors **R**ed, **G**reen **B**lue.

That is, image[3,10,2] gives you the value of blue for the pixel in the 3rd row and 10th column.

Each color value is given between 0 and 255:

```python
image.min(), image.max()
```

If we take the mean value of the 3 color channels, we get the grayscale image

```python
image_gray = image.mean(axis=2)
_ = plt.imshow(image_gray, cmap=plt.cm.gray)
```
